```python
from keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
from __future__ import print_function

import keras
import numpy as np
from keras.datasets import mnist
from keras.models import Model
from keras.layers import Dense, Dropout, Input
from keras.optimizers import RMSprop
```

```python
batch_size = 128
num_classes = 10
epochs = 20
```

```python
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')

# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
y_train[0]
```

```python
input_layer = Input(shape=(784,))
layer2 = Dense(512, activation='relu')(input_layer)
layer2 = Dropout(0.2)(layer2)
layer3 = Dense(512, activation='relu')(layer2)
layer3 = Dropout(0.2)(layer3)
output = Dense(num_classes, activation='relu')(layer3)

model = Model(input_layer, output)
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])
```

```python
x_train, x_valid = np.split(x_train, [55000])
y_train, y_valid = np.split(y_train, [55000])
```

```python
history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_valid, y_valid))
```

```python
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
